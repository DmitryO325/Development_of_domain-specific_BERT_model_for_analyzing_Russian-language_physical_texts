```python
from pprint import pprint
from transformers import pipeline
```

```python
'Английская модель'
classifier = pipeline("text-classification")  # подтягивается английская модель
print(classifier("Physics is science"))
```

```python
'Русская модель'

classifier = pipeline("text-classification", model='ai-forever/ruBERT-base')
print(classifier("Физика - лженаука"))  

'Модель не обучена под классификацию текстов'
```

```python
embedder = pipeline("feature-extraction", model='ai-forever/ruBERT-base')
print(embedder("Физика - лженаука"))  

'Можно получить эмбеддинги'
```

```python
unmasker = pipeline("fill-mask", model='ai-forever/ruBERT-base')
pprint(unmasker("Физика - хорошая наука, но некоторые считают, что она [MASK]."))  

'Можно предсказать пропущенное слово'
```

```python
ner = pipeline("ner", model='ai-forever/ruBERT-base')
pprint(ner("Физика - хорошая наука, но некоторые считают, что она бесполезная."))

'Модель не обучена на выделение именованных сущностей'
```

```python
classifier = pipeline("zero-shot-classification", model='ai-forever/ruBERT-base')
pprint(classifier(
    "Физика - хорошая наука, но некоторые считают, что она бесполезная.",
    candidate_labels=['физика', 'биология', 'математика', 'русский язык']
))

'Модель равновероятно относит текст про физику к этим 4 категориям'
```

```python
generator = pipeline('text-generation', model='ai-forever/ruBERT-base')
print(generator("Физика безусловно интересная наука, но программисту она"))

'Модель не обучена на генерацию, что логично'
```
